```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
import geopandas as gpd
import matplotlib.style as style
from matplotlib.backends.backend_pdf import PdfPages
```

```python
df = pd.read_csv("share-of-individuals-using-the-internet.csv")
```

```python
df.rename(columns={'Entity': 'Country', 'Code': 'Country_Code', 'Individuals using the Internet (% of population)': '% of Population Internet Users'}, inplace=True)
```

```python
df
```

```python
pdf_file = "2022-02-12_Carlson_Customizing Your Charts by Audience.pdf"
pp = PdfPages(pdf_file)
```

```python
def to_percent(y, position):
    # Transform y into an integer string
    s = str(int(round(1 * y)))
    return s + "%"
```

```python
fig, ax = plt.subplots(figsize=(10,9))

df_country = df[df['Country'] == 'United States']
plt.plot(df_country['Year'], df_country['% of Population Internet Users'], color='#16A3E9', linewidth=2, marker='o', label='USA')

df_china = df[df['Country'] == 'China']
plt.plot(df_china['Year'], df_china['% of Population Internet Users'], color='#E95C16', linewidth=2, marker='o', dashes=[.45], label='China')

plt.suptitle('Yearly Internet Usage Gap Between USA and China Closes Rapidly', fontname='Segoe UI', fontsize=18, ha='left', va='top', x=0.035, y=1.03)
plt.text(0.035, .92, "Internet adoption in the USA and China has grown significantly in the past 20 years, with the USA\nhaving a higher percentage of internet users. China has had rapid growth, reaching over 70%\ninternet access, while the USA has reached over 90%. The increasing access to the internet in\nboth countries highlights its growing importance."
         , ha='left', va='center', fontsize=14, color = '#707377',fontfamily='sans-serif', transform=plt.gcf().transFigure,
         linespacing=2)
plt.text(0.055, 0.01, "Source: ourworldindata.org", ha='left', va='center'
         , fontsize=12, color = 'black',fontname='Segoe UI'
         , transform=plt.gcf().transFigure,
         linespacing=1.5)

plt.gca().yaxis.grid(True, color='#B7C6CF')
plt.gca().xaxis.grid(False)

plt.ylim(0, 100)

plt.tight_layout(rect=[0, 0.05, 1, 1])

plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)

formatter = ticker.FuncFormatter(to_percent)
plt.gca().yaxis.set_major_formatter(formatter)


plt.legend(loc='upper center', ncol=2, fontsize = 14, frameon=False)

plt.show()
pp.savefig(fig)
```
